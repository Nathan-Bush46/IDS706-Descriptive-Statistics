```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
counts = {"apple": 6, "bannan": 4}
total_words = sum(counts.values())
p_b_values = np.linspace(0, 1, 100)  # p_b values from 0 to 1
p_a_values = 1 - p_b_values  # p_a = 1 - p_b

# Calculate the probability of the observations
probabilities = [p_a**counts["apple"] * p_b**counts["bannan"] for p_a, p_b in zip(p_a_values, p_b_values)]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(p_a_values, probabilities, label='Probability of Observations')
plt.xlabel('p_a')
plt.ylabel('Probability')
plt.title('Probability of Observations under the Unigram Model as a Function of p_a')
plt.grid(True)
plt.legend()
plt.show()
```
